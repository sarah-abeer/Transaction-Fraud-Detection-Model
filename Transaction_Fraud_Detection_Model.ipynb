```python
#distribution of legit transaction & fraudulent transaction
credit_card_data['isFraud'].value_counts()
```

0--> Normal Transaction
1-->Fraudlent Transaction

```python
#seprating the data for analysis
legit=credit_card_data[credit_card_data.isFraud==0]
fraud=credit_card_data[credit_card_data.isFraud==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-Sampling



Build a sample dataset containing similar distribution and Fraudulent Transaction

Number of Fraudulent Transaction--> 458

```python
legit_sample= legit.sample(n=458)
```

Concatenating two DataFrames

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```
